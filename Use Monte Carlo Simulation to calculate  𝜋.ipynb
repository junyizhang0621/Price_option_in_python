### let's use Monte Carlo Simulation to calculate $\pi$

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
def estimate_pi(N_sim=10000):
    M=0
    i=0
    while i < N_sim:
        rv1 = np.random.rand()
        x = 2*rv1-1
        rv2 = np.random.rand()
        y = 2*rv2-1

        #print(rv1,rv2)_
        if x**2 + y**2 < 1:
           M = M+1
        i=i+1

    pi_approx = M/N_sim*4
    return pi_approx
```

```python
estimate_pi(100)
```

```python
log_n = np.array([1,1.5,2,2.5,3,3.5,4,4.5,5,5.5,6])
n= 10**log_n
pi_est = np.zeros_like(log_n)
for i in range(len(log_n)):
    pi_est[i] = estimate_pi(n[i])
plt.plot(log_n, pi_est)
plt.scatter(log_n, pi_est)
plt.xlabel('Simulation times (log)')
plt.ylabel('Approximated pi value')
plt.axhline(np.pi, linestyle='--',color='red')
plt.title('Approximation of Pi via Monte Carlo')
```
